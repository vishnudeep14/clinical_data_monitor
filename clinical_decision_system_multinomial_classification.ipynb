```python

one_hot_encoded_data = pd.get_dummies(df, columns = ['prognosis'])
print(one_hot_encoded_data)
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
x=df.drop(["prognosis"],axis=1)
y=df["prognosis"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
mod=LogisticRegression(solver='lbfgs',multi_class='multinomial')
mod.fit(x_train, y_train)
mod.score(x_test, y_test)
```

```python
y_pred = mod.predict(x_test)
```

```python
real_dis=y_test.values
```

```python
for i in range(0, len(real_dis)):
    if y_pred[i] == real_dis[i]:
        print ('Pred: {0} Actual:{1}'.format(y_pred[i], real_dis[i]))
    else:
        print('worng prediction')
        print ('Pred: {0} Actual:{1}'.format(y_pred[i], real_dis[i]))
```
